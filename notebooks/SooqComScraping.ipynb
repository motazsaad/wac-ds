```
dev_name = [item.text.strip() for item in soup.find_all('h1', {'class': 'itemTitle'})]
print(len(dev_name))
dev_name
```

```
dev_price = [item.text.strip() for item in soup.find_all('h3', {'class': 'itemPrice'})]
print(len(dev_price))
dev_price
```

```
dev_name[6], dev_price[6]
```

```
dev_specs = [item.text for item in soup.find_all('ul', {'class': 'selling-points'})]
print(len(dev_specs))
dev_specs
```

```
dev_storage = [spec.strip().split('\n')[0].replace('سعة التخزين: ', '') for spec in dev_specs]
dev_screen = [spec.strip().split('\n')[1].replace('حجم الشاشة: ', '') for spec in dev_specs]
dev_os = [spec.strip().split('\n')[2].replace('نظام التشغيل: ', '') for spec in dev_specs]
dev_ram = [spec.strip().split('\n')[3].replace('حجم الذاكرة: ', '') for spec in dev_specs]
```

```
import pandas as pd 
tablet_df = pd.DataFrame({
    'name': dev_name,
    'price': dev_price, 
    'storage': dev_storage,
    'os': dev_os, 
    'screen': dev_screen,
    'ram': dev_ram
})
tablet_df
```
